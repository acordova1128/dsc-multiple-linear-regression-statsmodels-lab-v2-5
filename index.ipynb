## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(ncols=3, figsize=(15,5), sharey=True)
axes[0].set_ylabel("SalePrice")

for i, col in enumerate(ames_subset.drop("SalePrice", axis=1).columns):
    ax = axes[i]
    ax.scatter(ames_subset[col], ames_subset["SalePrice"], alpha=0.2)
    ax.set_xlabel(col)
````

````python
# Your written answer here - do these seem like good candidates for linear regression?
# All three distributions seem to be linear, with 1stFlrSF having the most variability.
# However, all three do have outliers that could affect the results.
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
ames_subset.corr()["SalePrice"]
y = ames_subset['SalePrice']
X_baseline = ames_subset[["GrLivArea"]]
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

baseline_model = sm.OLS(y, sm.add_constant(X_baseline))
baseline_results = baseline_model.fit()

print(baseline_results.summary())
````

````python
# Your written answer here - interpret model results
# The model has an R-squared value of .502, which means that it accounts for only about 50% of all the variance in the data.
# The constant coefficient is about 18570, which means for a home with 0 square feet of above-ground living area would cost about $18.6k.
# The constant coefficient is statistically significant as the p-value is less than alpha equal to 0.05.
# The GrLivArea coefficient is 107.1304, which means for every 1 square foot of ground floor living area, the price of a house increases $107 approximately.
# The GrLivArea coefficient is also statistically significant as the p-value is less than the above alpha level.
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X = ames_subset.drop("SalePrice", axis=1)
````

````python
# Your code here - fit model and display results
subset_model = sm.OLS(y, sm.add_constant(X))
subset_results = subset_model.fit()

print(subset_results.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?
# The model has an R-squared value of .565, which means that it accounts for 56.5% of all the variance in the data, which is higher than our base model.
# The constant coefficient is about -14310, which means for a home with 0 square feet of above-ground living area would cost -$14310.
# The constant coefficient is statistically significant as the p-value is less than alpha equal to 0.05.
# The LotArea coefficient is 0.2841, which means for every 1 square foot of ground floor living area, the price of a house increases by less than a dollar.
# The LotArea coefficient is not statistically significant as the p-value is greater than the above alpha level.
# The 1stFlrSF coefficient is 60.2866, which means for every 1 square foot of ground floor living area, the price of a house increases $60 approximately.
# The 1stFlrSF coefficient is statistically significant as the p-value is less than the above alpha level.
# The GrLivArea coefficient is 80.6061, which means for every 1 square foot of ground floor living area, the price of a house increases $81 approximately.
# The GrLivArea coefficient is also statistically significant as the p-value is less than the above alpha level.

# Overall, this model does have a higher R-squared value than our base model, however, one of our variables is not statistically significant; therefore
# the "better model" is case dependent.
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
fig = plt.figure(figsize=(15,5))
sm.graphics.plot_partregress_grid(
    subset_results,
    exog_idx=list(X.columns),
    grid=(1,3),
    fig=fig)
plt.show()
````

````python
# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?

# In general, 1stFlrSF and GrLivArea look about the same as they did in the scatter plot, but with less steep slopes.
# We can see for LotArea, the outliers affect the model more than we saw in the scatterplot, and overall, LotArea was much weaker as a predictor than it looked like initially.
# There's also not much overlap between the variance that 1stFlrSF and GrLivArea explain, even though they relate to square footage.
# It seems they must contain different enough information to affect the variance differently.
````
